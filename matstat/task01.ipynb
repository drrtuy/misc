# Задание 1 из ДЗ. Построить эмпирическую функцию распределения(empirical CDF)

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Задаю выборку
sample = np.array([25, 29, 19, 26, 23, 16, 20, 22, 24, 18, 18, 30, 26, 24, 24, 19])

ecdf = sm.distributions.ECDF(sample)

x = np.linspace(min(sample), max(sample))
y = ecdf(x)

fig = plt.figure()
cdf1 = fig.add_subplot(111)
cdf1.set_ylabel('кумулятивная частота')
cdf1.set_title('значение')
plt.step(x,y)
```

```python
# Можно получить частоту значений, вызвав объект ecdf.
ecdf([19,23])
```

# Задание 2 из ДЗ. Построить гистограмму

```python
# Задаем выборку
freq = [24, 25, 23, 27, 26, 25, 25, 23, 24, 23, 23, 26]
x_min = 3
x_max = 15
size = 294
i = 0
run = []
for i in range(3,x_max):
    run += [ i for v in range(0,freq[i-x_min]) ]

len(run) == size
```

# Раcпределение Chi^2


```python
fig2 = plt.figure()
hist1 = fig2.add_subplot(111)
hist1.set_ylabel('частота')
hist1.set_title('значение')

plt.hist(run, len(freq),
         density=True,
         histtype='bar',
         facecolor='b',
         alpha=0.5)

plt.show()
```
